```python
import pandas as pb
data ={'id':[1,2,3,4,5,6,7,8],
       'name':['banty','chanty','heymansh','shva','deepak','mani','got','kandakurthi'],
       'age':[21,32,20,25,44,23,19,99]
       }
```

```python
df = pb.DataFrame(data)
df
```

```python
#VIEWING OF DATA using  HEAD
#NEGTIVE VALUES WILL NOT ACCEPT
print(df.head())
print()
print(df.head(2))
```

```python
#VIEWING OF DATA using  TAIL
print(df.tail())
print()
print(df.tail(2))
```

```python
# DISPLAYING DATAFRAME INFORMATION
df.info()
```

```python
#discribing the dataframe
df.describe()
```

```python
#selecting a colum

df['age']
```

```python
#selecting multiple colums
df[['age','id']]
```

```python
#selecting rows by index
# if you dont know id of a data and you need to select the row then it is used
df.iloc[2]
```

```python
#SELECTING ROW BY CONDITION
df[df['age']>70]
```

```python
#select row by multiple conditions

df[(df['id']<10) & (df['age']>20)]
```

```python
#to add collum
#if collum is there it will update if not there it will creat

df['address']=['a','b','c','d','e','f','g','h']
df
```

```python
#using Aggricate functions
df['salary']=[234,234678,568,7654,8765,8765,46,3456]

df['total_salart'] = df['salary'].sum()
df
```

```python
#applying more than one aggrate function
df['salary'].agg(['min','max','mean','median'])
```
