```python
# if you dont have an headder

import pandas as pd

collums = ['id','name','salary','gender','dno']
c = pd.read_csv('/content/pandas_practiv_WITHOUT_HIDDING.txt', header=None , names=collums)
c
```

```python
#to replace the names

# task 1: in gender column, replace 'm' as 'Male', 'f' as 'Female'

# 3 ways.

# 1. replace()

# 2. map()

#3. apply()
#Use strip() along with lower() to handle extra spaces + case issues
```

```python
# using replace( )
# for the replace  input should be dictionary
#Use strip() along with lower() to handle extra spaces + case issues

c['gender']=c['gender'].str.lower().str.strip().replace({'m': 'Male', 'f': 'Female'})
c
```

```python
# using map( )

#*for the map also input should be dictionary.
#*If key not available it returns null.. It is difference Between replace and map
#
#astype(str) ‚Üí ensure string

#str.strip() ‚Üí remove spaces

#astype(int) ‚Üí convert back to integer

#map(dept) ‚Üí map department names correctly

import pandas as pd

dept = {11: 'marketing', 12: 'hr', 13: 'finance'}

c['dno'] = c['dno'].astype(str).str.strip().astype(int).map(dept)
c
```

```python

c = pd.read_csv('/content/pandas_practiv_WITHOUT_HIDDING.txt', header=None , names=['id','name','salary','gender','dno'])
c
```

```python
# USING APPLY()

# USING PYTHON FUNCTION CUSTOMISE PYTHON FUNCTIONS ON DATAFRAME

def n(g):
  g = g.lower().strip()
  gnder ='invalid'
  if g =='m':
    gnder = 'Male'
  elif g == 'f':
    gnder = 'Female'
  return gnder

c['gender']= c['gender'].str.strip().apply(n)
c
```

```python
def dnoo(n):
  department = 'invalid'
  if n== 11:
    department = 'marketing'
  elif n == 12:
    department = 'hr'
  elif n == 13:
    department = 'finance'
  return department

c['dno'] = c['dno'].apply(dnoo)
c
```
